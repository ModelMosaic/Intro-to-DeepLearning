# Introduction to Deep Learning

### Import the necessary libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
```

### Define a simple dataset

- We'll create a dataset of 2D points, where we'll classify them as either 'Red' or 'Blue'.
- This is a supervised learning problem, where we have labeled data.
- This dataset can be thought of as points on a coordinate plane, where we want to draw a line to separate Red and Blue points.


```python
# Generate random data points
np.random.seed(0)
X = np.random.randn(200, 2)  # 200 data points with 2 features (x1 and x2)
y = (X[:, 0] + X[:, 1] > 1).astype(int)  # Labeling: 1 if x1 + x2 > 1 else 0
```

### Visualize the data


```python
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Spectral)
plt.title("Simple Dataset (Red vs. Blue)")
plt.xlabel("Feature x1")
plt.ylabel("Feature x2")
plt.show()
```

### Define the model using TensorFlow/Keras to classify this data

```python
model = keras.Sequential([
    keras.layers.Input(shape=(2,)),           # Input layer with 2 features
    keras.layers.Dense(4, activation='relu'),  # Hidden layer with 4 neurons and ReLU activation
    keras.layers.Dense(1, activation='sigmoid')  # Output layer with 1 neuron and sigmoid activation (binary classification)
])
```

### Compile the model


```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

### Train the model

```python
history = model.fit(X, y, epochs=100, verbose=0)
```

### Plot the training loss and accuracy over epochs

```python
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(history.history['loss'])
plt.title("Training Loss")
plt.xlabel("Epochs")
plt.ylabel("Loss")

plt.subplot(1, 2, 2)
plt.plot(history.history['accuracy'])
plt.title("Training Accuracy")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")

plt.tight_layout()
plt.show()
```

### Make predictions

```python
# Generate new data points for prediction
new_data_points = np.array([[1.0, 1.0], [-1.0, -1.0], [0.0, 0.0]])

# Predict the class labels for the new data points
predictions = model.predict(new_data_points)

# Print the predictions
for i, prediction in enumerate(predictions):
    class_label = 'Red' if prediction < 0.5 else 'Blue'
    print(f"Data point {i+1} is predicted as '{class_label}'")
```

### Explanation

- We created a simple dataset with two features (x1 and x2) and labeled the data points as Red or Blue based on a condition.
- We visualized the data points to understand the problem.
- We defined a neural network model using TensorFlow/Keras with an input layer, a hidden layer, and an output layer.
- We compiled the model by specifying the optimizer, loss function, and evaluation metrics.
- We trained the model using the dataset and visualized the training progress.
- Finally, we used the trained model to make predictions on new data points.
